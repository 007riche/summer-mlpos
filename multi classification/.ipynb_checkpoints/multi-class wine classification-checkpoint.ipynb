```python
# encode y values into dummy Vars (using pandas)
y_cat = pd.get_dummies(y)
```

```python
X = dataset.drop('Class', axis=1 # to perform column wise operation on pandas Obj
                )
```

```python
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
from keras.layers import Dense 
```

```python
model.get_config()
```

```python
model.add(Dense(
    units=5,   #No. of neurons
    input_shape=(13,),  # input shape
    activation='relu',
    kernel_initializer='he_normal'
))
```

```python
model.summary()
```

```python
model.add(Dense(
    units=8,   #No. of neurons
    activation='relu',
    kernel_initializer='he_normal'
))
```

```python
model.add(Dense(
    units=8,   #No. of neurons
    activation='relu',
    kernel_initializer='he_normal'
))
```

```python
model.summary()
```

```python
# Output layer
model.add(Dense(
    units=3,   #No. of neurons based on Number of classes(Target)
    activation='softmax', # probabilistic mutliclass activation
    
))
```

```python
model.summary()
```

```python
model.get_config()
```

```python
# Optimizer importation
from keras.optimizers import RMSprop
```

```python
model.compile(optimizer=RMSprop(learning_rate=0.01), 
              loss='categorical_crossentropy', # Probabilistic multiclass loss function
              metrics=['accuracy'] # To display the accuray, Not a hyper parameter
             )
```

```python
model.fit(X, y_cat, epochs=5000)
```

```python
model.history.history
```

```python
hist = model.history.history['loss']
```

```python
type(hist)
```

```python
# Conversion into pandas datatpye, DataFrame datatype
hist_pd= pd.DataFrame(hist)
```

```python
# Plotting
hist_pd.plot()
```

```python
acuracy = model.history.history['accuracy']
```

```python
acur_pd = pd.DataFrame(acuracy)
```

```python
acur_pd.plot()
```

```python
model.save('wineClassifier.h5')
```
