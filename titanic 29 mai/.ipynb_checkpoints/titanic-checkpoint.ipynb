```python
#HERE, we will use the average age with respect to the Pclass
```

```python
import pandas as pd
```

```python
def fillMissingVals(colTuple):
    Pclass=colTuple[0]
    age=colTuple[1]
    
    if pd.isnull(age):
        if Pclass == 1:
            return 38
        elif Pclass == 2:
            return 30
        elif Pclass == 3:
            return 25
        else:
            return 30
    else:
        return age
    
```

```python
# apply method here is a Pandas method that helps to apply a function, for instance on selected data here
dataset[["Pclass", "Age"]].apply(fillMissingVals, axis=1)
```

```python
#Updating
age=dataset[["Pclass", "Age"]].apply(fillMissingVals, axis=1)
```

```python
X['Age']=age
```
