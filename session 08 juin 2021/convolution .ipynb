```python
from scipy import misc
```

```python
provide_img=misc.ascent()
```

```python
provide_img.shape
```

```python
mpl.imshow(provide_img)
```

```python
mpl.imshow(provide_img, cmap='gray')
```

```python
#create manually a filter
f= numpy.array([[1,2,1],
           [0,0,0],
           [-1,-2,-1]])
```

```python
from scipy.signal import convolve2d
```

```python
cimg= convolve2d(provide_img, f)
```

```python
cimg
```

```python
cimg.shape
```

```python
mpl.imshow(cimg, cmap='gray')
```

```python
f= numpy.array([[0,0,0],
           [0,-4,0],
           [0,0,0]])
```

```python
cimg= convolve2d(provide_img, f)
```

```python
mpl.imshow(cimg, cmap='gray')
```

```python
mpl.imshow(cimg)
```
